```python
from __future__ import print_function
from sklearn import datasets
from sklearn.linear_model import LinearRegression

loaded_data = datasets.load_boston()
data_X = loaded_data.data
data_y = loaded_data.target

model = LinearRegression()
model.fit(data_X, data_y)

print(model.predict(data_X[:4, :]))
#整个参数下面两个
#系数，每个属性配套的参数，房屋的13个参数
print(model.coef_)
#截距，常数项
print(model.intercept_)

#输出之前定义的参数
print(model.get_params())
#对学的东西打分，就是相关系数类似（0-1）
print(model.score(data_X, data_y)) # R^2 coefficient（系数）of determination（测定）
```

```python
from __future__ import print_function
from sklearn import preprocessing #预处理的意思
import numpy as np
#区分训练集和测试集
from sklearn.model_selection import train_test_split
#生成一些用于分类的数据集
from sklearn.datasets.samples_generator import make_classification
#支持向量机，大间距分类器
from sklearn.svm import SVC
#显示
import matplotlib.pyplot as plt

a = np.array([[10, 2.7, 3.6],
                     [-100, 5, -2],
                     [120, 20, 40]], dtype=np.float64)
#对比scale的作用，对列向量进行处理，偏差为0，方差为1
print(a)
#下面是标准化之后的结果，更适合去处理ML
print(preprocessing.scale(a))
#数量，属性，冗余，信息，随机状态，
X, y = make_classification(n_samples=300, n_features=2 , n_redundant=0, n_informative=2,
                           random_state=22, n_clusters_per_class=1, scale=100)
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
X = preprocessing.scale(X)    # normalization 十分重要
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3)
clf = SVC()
#训练集拟合
clf.fit(X_train, y_train)
#测试集打分
print(clf.score(X_test, y_test))
```

```python
# View more python learning tutorial on my Youtube and Youku channel!!!

# Youtube video tutorial: https://www.youtube.com/channel/UCdyjiB5H8Pu7aDTNVXTTpcg
# Youku video tutorial: http://i.youku.com/pythontutorial

"""
Please note, this code is only for python 3+. If you are using python 2+, please modify the code accordingly.
"""
from __future__ import print_function
from sklearn.datasets import load_iris
#交叉验证
from sklearn.cross_validation import train_test_split
#一种分类器model
from sklearn.neighbors import KNeighborsClassifier

iris = load_iris()
X = iris.data
y = iris.target

# test train split #分开测试集与训练集，5组不同范围测试集
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=4)
#考虑到5个邻居
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
print(knn.score(X_test, y_test))#测试集打分0.97

# this is cross_val_score #
from sklearn.cross_validation import cross_val_score
knn = KNeighborsClassifier(n_neighbors=5)
#共5组，每一组的test，train不同，平均更准确
scores = cross_val_score(knn, X, y, cv=5, scoring='accuracy')
print(scores)

# this is how to use cross_val_score to choose model and configs #
from sklearn.cross_validation import cross_val_score
import matplotlib.pyplot as plt
k_range = range(1, 31)
k_scores = []
for k in k_range:#寻找最佳k，少了欠拟合，多了过拟和
    knn = KNeighborsClassifier(n_neighbors=k)
##    loss = -cross_val_score(knn, X, y, cv=10, scoring='mean_squared_error') # for regression回归问题
    scores = cross_val_score(knn, X, y, cv=10, scoring='accuracy') # for classification分类问题
    k_scores.append(scores.mean())

plt.plot(k_range, k_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Accuracy')
plt.show()

```

```python
from __future__ import print_function
#学习曲线，可视化
from sklearn.learning_curve import  learning_curve
#搞数据
from sklearn.datasets import load_digits
#model支持向量机
from sklearn.svm import SVC
import matplotlib.pyplot as plt
import numpy as np

digits = load_digits()
X = digits.data
y = digits.target
train_sizes, train_loss, test_loss= learning_curve(
        SVC(gamma=0.001), X, y, cv=10, scoring='neg_mean_squared_error',
        train_sizes=[0.1, 0.25, 0.5, 0.75, 1])
train_loss_mean = -np.mean(train_loss, axis=1)
test_loss_mean = -np.mean(test_loss, axis=1)

#训练
plt.plot(train_sizes, train_loss_mean, 'o-', color="r",label="Training")
#测试
plt.plot(train_sizes, test_loss_mean, 'o-', color="g",label="Cross-validation")

plt.xlabel("Training examples")
plt.ylabel("Loss")
plt.legend(loc="best")
plt.show()
```

```python
# View more python learning tutorial on my Youtube and Youku channel!!!

# Youtube video tutorial: https://www.youtube.com/channel/UCdyjiB5H8Pu7aDTNVXTTpcg
# Youku video tutorial: http://i.youku.com/pythontutorial

"""
Please note, this code is only for python 3+. If you are using python 2+, please modify the code accordingly.
"""
from __future__ import print_function
from sklearn.learning_curve import  validation_curve
from sklearn.datasets import load_digits
from sklearn.svm import SVC
import matplotlib.pyplot as plt
import numpy as np

digits = load_digits()
X = digits.data
y = digits.target
#参数的范围，通过log，5个点
param_range = np.logspace(-6, -2.3, 5)
#改变gamma
train_loss, test_loss = validation_curve(
        SVC(), X, y, param_name='gamma', param_range=param_range, cv=10,
        scoring='neg_mean_squared_error')
train_loss_mean = -np.mean(train_loss, axis=1)
test_loss_mean = -np.mean(test_loss, axis=1)

plt.plot(param_range, train_loss_mean, 'o-', color="r",
             label="Training")
plt.plot(param_range, test_loss_mean, 'o-', color="g",
             label="Cross-validation")
#验证曲线，寻找最佳gamma，右边是过拟合，现在的验证基本上都是交叉验证
plt.xlabel("gamma")
plt.ylabel("Loss")
plt.legend(loc="best")
plt.show()
```
